```julia
using Distributions, TimeIt, ProgressMeter, PyPlot, QuadGK
```

```julia
mu_1, mu_2 = -2, 2
sigma = 1
function f(x, temp) 
    return (0.5/(sqrt(2*pi)*sigma)*exp.(-(x-mu_1).^2/(2*sigma^2)) + 
        0.5/(sqrt(2*pi)*sigma)*exp.(-(x-mu_2).^2/(2*sigma^2))).^temp
end
```

```julia
temp_list = linspace(0.1,1,10)
int_list = zeros(length(temp_list))
for (i,temp) in enumerate(temp_list)
    function f(x) 
        return (0.5/(sqrt(2*pi)*sigma)*exp.(-(x-mu_1).^2/(2*sigma^2)) + 
            0.5/(sqrt(2*pi)*sigma)*exp.(-(x-mu_2).^2/(2*sigma^2))).^temp
    end
    int_list[i] = quadgk(f, -15, 15, rtol=1e-8)[1]
end
```

```julia
fig = figure(figsize=(6,3))
xgrid = linspace(-15,15,1000)

function f(x, temp) 
    return (0.5/(sqrt(2*pi)*sigma)*exp.(-(x-mu_1).^2/(2*sigma^2)) + 
        0.5/(sqrt(2*pi)*sigma)*exp.(-(x-mu_2).^2/(2*sigma^2))).^temp
end

for (i,temp) in enumerate(temp_list)
    plot(xgrid, f(xgrid, temp)/int_list[i], label="s = "*string(i))
end
xlabel("x")
ylabel("pi_s(x)")
title("Density")
grid(true)
xlim(-15,15)
legend(ncol=5,bbox_to_anchor=(1.05, 1.35))
savefig("plots/temps.pdf", format="pdf", dpi=10000, bbox_inches="tight");
```
