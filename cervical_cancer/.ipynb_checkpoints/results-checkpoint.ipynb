```julia
using Distributions, TimeIt, PyPlot, JLD, ProgressMeter, CSV
include("/home/postdoc/dsen/Desktop/Biometrika-code/zz_structures.jl")
include("/home/postdoc/dsen/Desktop/Biometrika-code/mbsampler.jl")
```

```julia
n_samples = 10^4

filename = "/xtmp/PDMP_data_revision/cervical_cancer/uniform_CV.jld"
skel, bt = load(filename, "xt_skeleton", "bt_skeleton")
xi_samples_uniform_CV = extract_samples(skel, bt, bt[end]/n_samples, "linear")

filename = "/xtmp/PDMP_data_revision/cervical_cancer/uniform_noCV.jld"
skel, bt = load(filename, "xt_skeleton", "bt_skeleton")
xi_samples_uniform_noCV = extract_samples(skel, bt, bt[end]/n_samples, "linear")

filename = "/xtmp/PDMP_data_revision/cervical_cancer/wt_CV.jld"
skel, bt = load(filename, "xt_skeleton", "bt_skeleton")
xi_samples_wt_CV = extract_samples(skel, bt, bt[end]/n_samples, "linear")

filename = "/xtmp/PDMP_data_revision/cervical_cancer/wt_noCV.jld"
skel, bt = load(filename, "xt_skeleton", "bt_skeleton")
xi_samples_wt_noCV = extract_samples(skel, bt, bt[end]/n_samples, "linear")

filename = "/xtmp/PDMP_data_revision/cervical_cancer/SS.jld"
skel, bt = load(filename, "xt_skeleton", "bt_skeleton")
xi_samples_SS = extract_samples(skel, bt, bt[end]/n_samples, "linear");
```

```julia
maxlag = 1000
fig = figure(figsize=(15,2))
a, b = -0.4, 1.1

for i in 1:size(xi_samples_uniform_noCV,1)
    subplot(151)
    plot(acf(xi_samples_uniform_CV[i,:],maxlag), color="blue")
    grid(true)
    title("Uniform subsampling \n (with CV)")
    ylim(a,b)
    xlabel("Lag")
    ylabel("ACF")
    
    subplot(152)
    plot(acf(xi_samples_uniform_noCV[i,:],maxlag), color="blue")
    grid(true)
    title("Uniform subsampling \n (no CV)")
    ylim(a,b)
    xlabel("Lag")
    yticks(alpha=0)
    
    subplot(153)
    plot(acf(xi_samples_wt_CV[i,:],maxlag), color="blue")
    grid(true)
    title("Weighted subsampling \n (with CV)")
    ylim(a,b)
    xlabel("Lag")
    yticks(alpha=0)
    
    subplot(154)
    plot(acf(xi_samples_wt_noCV[i,:],maxlag), color="blue")
    grid(true)
    title("Weighted subsampling \n (no CV)")
    ylim(a,b)
    xlabel("Lag")
    yticks(alpha=0)
    
    subplot(155)
    plot(acf(xi_samples_SS[i,:],maxlag), color="blue")
    grid(true)
    title("Stratified subsampling")
    ylim(a,b)
    xlabel("Lag")
    yticks(alpha=0)
end
subplots_adjust(wspace=0.01)
```

```julia
d = size(xi_samples_uniform_noCV,1)
acf_uniform_CV = zeros(d, maxlag+1)
acf_uniform_noCV = zeros(d, maxlag+1)
acf_wt_CV = zeros(d, maxlag+1)
acf_wt_noCV = zeros(d, maxlag+1)
acf_SS = zeros(d, maxlag+1)

for i in 1:d 
    acf_uniform_CV[i,:] = acf(xi_samples_uniform_CV[i,:],maxlag)
    acf_uniform_noCV[i,:] = acf(xi_samples_uniform_noCV[i,:],maxlag)
    acf_wt_CV[i,:] = acf(xi_samples_wt_CV[i,:],maxlag)
    acf_wt_noCV[i,:] = acf(xi_samples_wt_noCV[i,:],maxlag)
    acf_SS[i,:] = acf(xi_samples_SS[i,:],maxlag)
end
```

```julia
print(round(mean(sum(acf_uniform_CV,2)),1), "\n")
print(round(mean(sum(acf_uniform_noCV,2)),1), "\n")
print(round(mean(sum(acf_wt_CV,2)),1), "\n")
print(round(mean(sum(acf_wt_noCV,2)),1), "\n")
print(round(mean(sum(acf_SS,2)),1))
```
