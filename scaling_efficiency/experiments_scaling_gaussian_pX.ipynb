### Scaling with pX:

```julia
mb_size = 1
pX_list = logspace(-2,0,8)
Nobs_list = repeat([500], length(pX_list))
attempts = repeat([10^5], length(pX_list))
cov_dist = "Gaussian"
n_reps = 50
varying = "pX"
include_CV, include_stratified = true, false
prior_sigma = 1e5;
```

```julia
start = time()
for rep in 1:n_reps 
    for i in 1:length(pX_list) 
        my_model, root = generate_model(5, Nobs_list[i], pX_list[i], cov_dist, prior_sigma) 
        run_mbsamplers(my_model, root, mb_size, attempts[i], pX_list[i], 
                       Nobs_list[i], varying, rep, cov_dist, include_CV, include_stratified)
    end
end
print("\n Total time = ", round((time()-start)/60,2), " mins")
```
